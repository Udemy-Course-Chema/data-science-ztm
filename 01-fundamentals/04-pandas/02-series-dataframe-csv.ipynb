# Series, DataFrames & CSV

```python
import pandas as pd
```

```python
# 2 main datatypes
series = pd.Series(["BMW","Toyota","Mercedes Benz"])
series
```

```python
# Series = 1-dimensional
colours = pd.Series(["Red","Blue","White"])
colours
```

```python
# DataFrame = 2-dimensional
car_data = pd.DataFrame({"Car_make": series, "Colors":colours})
car_data
```

```python
# Import Data
car_sales = pd.read_csv('./../DataFrames/car-sales.csv')
car_sales
```

```python
# Exporting a DataFrame
car_sales.to_csv('./../DataFrames/exportando.csv', index=False)
#car_sales.to_csv('./../DataFrames/exportando.csv')
```
